```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Load data
iris = pd.read_csv("IRIS.csv")
x = iris.iloc[:, :-1].values
y = iris.iloc[:, -1].values

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(x)

# Plot KMeans clusters
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s=50, c='green', label='Cluster 2')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s=50, c='blue', label='Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=150, c='yellow', label='Centroids', marker='X')
plt.title("KMeans Clustering")
plt.legend()

# Plot dendrogram
plt.subplot(1, 2, 2)
dendrogram(linkage(x, method='ward'))
plt.title("Hierarchical Clustering Dendrogram")
plt.tight_layout()
plt.show()
```
