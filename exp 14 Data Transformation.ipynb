```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
data = pd.read_csv(r"C:/Users/DEEPIKA PORIDA/Downloads/IRIS.csv")


import pandas as pd
from sklearn.preprocessing import RobustScaler, StandardScaler

# Load the dataset
df = pd.read_csv("C:/Users/DEEPIKA PORIDA/Downloads/IRIS.csv")

# Select numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# 1️⃣ Robust Scaling
robust_scaler = RobustScaler()
df_scaled = df.copy()
df_scaled[numeric_cols] = robust_scaler.fit_transform(df[numeric_cols])

# 2️⃣ Standard Scaling
standard_scaler = StandardScaler()
df_standardized = df.copy()
df_standardized[numeric_cols] = standard_scaler.fit_transform(df[numeric_cols])

# Print results
print("Robust Scaled Data (first 5 rows):\n", df_scaled[numeric_cols].head())
print("\nStandard Scaled Data (first 5 rows):\n", df_standardized[numeric_cols].head())
```
